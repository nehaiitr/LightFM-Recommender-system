```python
import pandas as pd
import numpy as np
```

```python
#dictionary mapping respective integers to product id
id2prod= products['id'].to_dict()
```

```python
#dictionary mapping product ids to respective integers
prod2id= {v: k for k, v in id2prod.items()}
```

```python
#function to clean dataframe text

def tokenize(str):
  str= str[1:-1]
  str= str.replace('\n','')
  import string
  exclude = set(string.punctuation)
  py_str = ''.join(ch for ch in str if ch not in exclude)
  str= py_str.replace(" ","")
  n = 10
  res = [str[i:i + n] for i in range(0, len(str), n)]

  return res
```

```python
sessions['prev_items']= sessions['prev_items'].apply(tokenize)
```

```python
sessions['next_item']= sessions['next_item'].map(prod2id)
```

```python
for i in range(0, 3606249):
  result = [prod2id[elem] for elem in sessions['prev_items'][i]]
  sessions['prev_items'][i] = result
```

```python
sessions
```

```python
from scipy import sparse
```

```python
#matrx to store user item interaction

from scipy.sparse import dok_matrix
matrix= dok_matrix((3606249, 1551057))
```

```python
for i in range(0, 3606249):
  for j in range(0, len(sessions['prev_items'][i])):
    matrix[i, sessions['prev_items'][i][j]]+= 1

matrix= matrix.tocsr()
```
